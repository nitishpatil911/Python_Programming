**While Loop**

```python
#while loop we can execute a set of statements as long as a condition is true.

i = 1
while i < 5:
  print(i)
  i= i + 1
```

```python
 # Print even Numbers from 0 to 10
i = 0
while i < 10:
  print(i)
  i= i + 2
```

```python
# Print Odd Numbers from 0 to 10

i = 1
while i < 10:
  print(i)
  i= i + 2
```

**Break Statement**

```python
#break statement we can stop the loop even if the while condition is true:
i = 1
while i < 10:
  print(i)
  if i == 4:
    break
  i += 1
```

**Continue Statement**

```python
#continue statement we can stop the current iteration, and continue with the next:
i = 0
while i < 10:
  i += 1
  if i == 5:
    continue
  elif i==8:
    continue
  print(i)
```

```python
# WAP to find sum of first n natural numbers.
n = int(input("enter the last number: "))
i = 1
sum = 0
while (i <= n):
    sum = sum + i
    i = i + 1
print("The sum is: ", sum)
```

**For Loop**

```python
#A for loop is used for iterating over a sequence
 #(that is either a list, a tuple, a dictionary or a string).
```

```python
L1 = ["A", "B", "C","D"]
for x in L1:
  print(x)
```

```python
for x in "AAFT":
  print(x)
```

```python
#break Statement
#Exit the loop when x is "B":


L1 = ["A", "B", "C","D"]
for x in L1:
  print(x)
  if x == "B":
    break
```

```python
#Exit the loop when x is "B", but this time the break comes
# before the print:
L1 = ["A", "B", "C","D"]
for x in L1:
  if x == "B":
    break
  print(x)
```

```python
#Continue Statement

L1 = ["A", "B", "C","D"]
for x in L1:
  if x == "B":
   continue
  print(x)
```

```python
#Nested  For Loops
L2 = [1,2,3]
L3 = ["Google", "Meta", "Fujitsu"]

for x in L2:
  for y in L3:
    print(x, y)
```

```python
# WAP for factorial of a number using for loop.

num = int(input("Enter any number:"))

# Initialize the factorial variable to 1
fact = 1

# Calculate the factorial using a for loop
for i in range(1, num + 1):
    fact =fact* i
print("Factorial of a number is",fact)
```

```python
# WAP for factorial of a number using while loop.

n = int(input("Enter any number: "))
# logic to calculate the factorial of a number
f = 1
while n >= 1:
    f = f * n
    n = n- 1
# print output
print("Factorial is", f)
```

```python
# WAP to check whether a given number is prime or not?
num = int(input("Enter a number: "))

if num == 0 or num == 1:
    print(num, "is not a prime number")
elif num > 1:
   # check for factors
   for i in range(2,num):
       if (num % i) == 0:
           print(num,"is not a prime number")
           break
   else:
       print(num,"is a prime number")

# if input number is less than
# or equal to 1, it is not prime
else:
   print(num,"is not a prime number")
```

```python
 #WAP to find reverse of a number?
num=int(input(" Enter any number:"))
rev=0
while(num>0):
  rev=(rev*10)+num%10
  num=num//10
print("Reverse of given number is:",rev)
```

```python
#WAP to check whether given number is Palindrome or not?
num=int(input(" Enter any number:"))
rev=0
x=num
while(num>0):
  rev=(rev*10)+num%10
  num=num//10
if(x==rev):
  print("Given Number is Palindrome")
else:
  print("Given number is not Palindrome")
```

```python
#WAP to check whether given number is Armstrong or not?
n=int(input("Enter number:"))
x=n
sum=0
while(n>0):
  sum=sum+(n%10)*(n%10)*(n%10)
  n=n//10
if x==sum:
  print("Number is Armstrong")
else:
  print("Number is not Armstrong")
```
