```python
car_dict={"brand":"BMW","model":"iX","year":1964}
print(car_dict)
```

**Dictionaries are used to store data values in key:value pairs.**

A dictionary is a collection which is ordered*, changeable and do not allow duplicates.

Dictionary items are presented in key:value pairs, and can be referred to by using the key name.

```python
#Accessing Dictionary elements
print(car_dict['brand'])
```

```python
# Length of Dictionary
print(len(car_dict))
```

```python
# Type of Data
print(type(car_dict))
```

```python
# For Duplicate keys
car_dict2={"brand":"BMW","model":"iX","year":1964,"year":2000}
```

```python
car_dict2
```

```python
car_dict3={"brand":"BMW","model":"iX","year":2024,"year":2000}
```

```python
#The keys() method will return a list of all the keys in the dictionary.
car_dict.keys()
```

```python
#The values() method will return a list of all the values in the dictionary.

car_dict.values()
```

```python
# To change vlaues of Dictionary
car_dict["year"]=2024
```

```python
car_dict
```

```python
# To add new key:value in dictionary
car_dict["color"]="White"
```

```python
#The values in dictionary items can be of any data type
d3={"name":"A",'year':2023,'Pass':True,'Marks':[50,60,70]}
```

```python
print(d3)
```

```python
d3.update({'year':2024})
```

```python
d3['phone_no']=94566
```

```python
# For Removing Dictionary items
#The pop() method removes the item with the specified key name:

d3.pop('phone_no')
```

```python
d3
```

```python
#The popitem() method removes the last inserted item
d3.popitem()
```

```python
d3
```

```python
d3.pop()
```

```python
#The del keyword removes the item with the specified key name:

del d3['Marks']
```

```python
#The del keyword can also delete the dictionary completely:
del d3
```

```python
#The clear() method empties the dictionary:
d3.clear()
```
