```python
# Functions

#Types of Arguments:

# Default Argument
# Keyword Argument
# Positional Argument

```

```python
def power(a,b):
    return a**b
```

```python
power(2,3)
```

```python
power(2)
```

```python
def power(a=1,b=1):
    return a**b              #default parameters
```

```python
power(2)
```

```python
power()
```

```python
# positional arguments
```

```python
power(2,3)
```

```python
power(3,2)
```

```python
#keyword argument
power(b=3,a=3)
```

```python
power(b=3,a=4)
```

```python
# *agrs & **kwargs

```

```python
def product(a,b,c):
    return a*b*c
```

```python
product(1,2,3)
```

```python
def multiply(*args):
     product=1
     for i in args:
       product=product*i
     return product
```

```python
multiply(1,2,3,4,5,6,7,8,9)
```
