```python
# Factorial Number Recursive Function

def fact(n):
  if n==0:
    return 1
  else:
    return n*fact(n-1)
fact(4)
```

```python
# Fibonacci Number using recursion

def fibonacci(n):
  if n < 0:
    raise ValueError("Input 0 or greater only!")
  fiblist = [0, 1]
  for i in range(2,n+1):
    fiblist.append(fiblist[i-1] + fiblist[i-2])
  return fiblist[n]

fibonacci(10)
```
